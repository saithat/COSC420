```python
import pandas as pd
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("results_v2.csv")
```

```python
#df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df
```

```python
N = [25, 50, 75, 100]
p_m = [0, 0.01, 0.03, 0.05]
p_c = [0, 0.1, 0.3, 0.5]
trn_size = [2, 3, 4, 5]
```

```python
colors = cm.rainbow(np.linspace(0, 1, 20))
colors_2 = cm.rainbow(np.linspace(0, 1, 16))
colors_lg = cm.rainbow(np.linspace(0, 1, 80))
```

```python
d_N = 50
d_p_m = 0.01
d_p_c = 0.3
d_tr_size = 2
```

# All default parameters

```python
index = 0
fig, ax = plt.subplots(figsize=(9,6))
for label, tmp_df in df[(df["N"]==d_N) & (df["p_m"]==d_p_m) & (df["p_c"]==d_p_c) & (df["tournament_size"]==d_tr_size)].groupby('iteration'):
    tmp_df.plot(x="generation", y="average_fitness", kind="line", ax=ax, xlabel="Generation", ylabel="Average Fitness", label=label, color=np.array([colors[index]]))
    index += 1
plt.legend(title="Iteration")
plt.title("Population size: {}, Mutation prob: {}, Crossover prob: {}, Tournament size: {}".format(d_N, d_p_m, d_p_c, d_tr_size))
```

```python
index = 0
fig, ax = plt.subplots(figsize=(9,6))
for label, tmp_df in df[(df["N"]==d_N) & (df["p_m"]==d_p_m) & (df["p_c"]==d_p_c) & (df["tournament_size"]==d_tr_size)].groupby('iteration'):
    tmp_df.plot(x="generation", y="best_fitness", kind="line", ax=ax, xlabel="Generation", ylabel="Best Fitness", label=label, color=np.array([colors[index]]))
    index += 1
plt.legend(title="Iteration")
plt.title("Population size: {}, Mutation prob: {}, Crossover prob: {}, Tournament size: {}".format(d_N, d_p_m, d_p_c, d_tr_size))
```

```python
index = 0
fig, ax = plt.subplots(figsize=(9,6))
for label, tmp_df in df[(df["N"]==d_N) & (df["p_m"]==d_p_m) & (df["p_c"]==d_p_c) & (df["tournament_size"]==d_tr_size)].groupby('iteration'):
    tmp_df.plot(x="generation", y="diversity_metric", kind="line", ax=ax, xlabel="Generation", ylabel="diversity_metric", label=label, color=np.array([colors[index]]))
    index += 1
plt.legend(title="Iteration")
plt.title("Population size: {}, Mutation prob: {}, Crossover prob: {}, Tournament size: {}".format(d_N, d_p_m, d_p_c, d_tr_size))
```

# Vary default - average fitness

```python
index = 0
fig, ax = plt.subplots(figsize=(20,10))
for n in N:
    for label, tmp_df in df[(df["N"]==n) & (df["p_m"]==d_p_m) & (df["p_c"]==d_p_c) & 
    (df["tournament_size"]==d_tr_size)].groupby('iteration'):
        tmp_df.plot(x="generation", y="average_fitness", 
        kind="line", ax=ax, label=str(n)+'_'+str(label), xlabel="Generation", ylabel="Average of Average Fitness", color=np.array([colors_lg[index]]))
        index += 1
plt.legend(title="Population size")
plt.title("Mutation prob: {}, Crossover prob: {}, Tournament size: {}".format(d_p_m, d_p_c, d_tr_size))
```

```python
index = 0
fig, ax = plt.subplots(figsize=(9,6))
for n in N:
    df[(df["N"]==n) & (df["p_m"]==d_p_m) & (df["p_c"]==d_p_c) & 
    (df["tournament_size"]==d_tr_size)].groupby('generation').mean().reset_index().plot(x="generation", y="average_fitness", 
    kind="line", ax=ax, label=n, xlabel="Generation", ylabel="Average of Average Fitness", color=np.array([colors[index]]))
    index += 1
plt.legend(title="Population size")
plt.title("Mutation prob: {}, Crossover prob: {}, Tournament size: {}".format(d_p_m, d_p_c, d_tr_size))
```
