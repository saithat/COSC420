```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
def traffic_ca(steps, size, rho):
    x = np.zeros((steps, size))
    x[0,:int(rho*size)] = 1
    np.random.shuffle(x[0, :])
    speeds = []
    
    lookup_table = [0, 0, 0, 1, 1, 1, 0, 1]
    
    for t in range(steps-1):
        for j in range(size):
            left = (j-1)%size
            right = (j+1)%size
            n_state = 4*x[t,left]+2*x[t,j]+x[t,right]
            x[t+1,j] = lookup_table[int(n_state)]
            if (t > 50 and x[t, j] == 1):
                if (x[t+1,j] == 1):
                    speeds.append(0)
                else:
                    speeds.append(1)
    
    return x
```

```python
steps = 100
size = 100
rho = 0.7 # Density of cars

x = traffic_ca(steps, size, rho)

plt.imshow(x, cmap='binary')
print(np.mean(speeds))
```
